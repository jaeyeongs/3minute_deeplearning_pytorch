**3.1.3 Autograd**

```python
import torch
```

```python
# 값이 1.0인 스칼라 텐서 정의
w = torch.tensor(1.0, requires_grad=True)  # requires_grad=True 지정하면 w에 대한 미분값을 w.grad에 저장
```

```python
a = w*3
```

```python
l = a**2
```

```python
# l을 w로 미분하기 위해 backward() 함수 사용
l.backward()
print('l을 w로 미분한 값은 {}'.format(w.grad))
```
