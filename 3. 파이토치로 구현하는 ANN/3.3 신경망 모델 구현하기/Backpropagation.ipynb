# Backpropagation (hard coding)

```python
import torch.nn.functional as F
import torch
```

```python
input = torch.FloatTensor([0.0, 1.0])
w1 = 3.34
b1 = -1.43
w2 = -3.53
b2 = 0.57
b3= 0
 
y1 = F.softplus(input*w1+b1)
y2 = F.softplus(input*w2+b2)
w3 = -1.22 
w4 = -2.30

predicted = y1 * w3 +  y2* w4 + b3

predicted
```

```python
# predicted value for an input
predicted[0].item()
```

```python
# loss for an input
-2*(0 - predicted[0].item())
```

```python
# SSR or slope
slope = sum([-2*(0 - predicted[0].item())])
slope
```

```python
# Gradient Descent
learning_rate = 0.1
step_size = slope * learning_rate
b3 = b3 - step_size
b3
```

```python
# Entire process of Gradient descent
input = torch.FloatTensor([0.0, 0.5, 1.0])
w1 = 3.34
b1 = -1.43
w2 = -3.53
b2 = 0.57
b3= 0
 
y1 = F.softplus(input*w1+b1)
y2 = F.softplus(input*w2+b2)
w3 = -1.22 
w4 = -2.30
observed = [0,1,0]

for i in range(15):  # 15 = num of steps
  print('observed = ',observed)
  predicted = y1 * w3 +  y2* w4 + b3

  print('predicted = ',predicted)

  slope = sum([-2*(observed[0] - predicted[0].item()),
               -2*(observed[1] - predicted[1].item()),
               -2*(observed[2] - predicted[2].item())])
  print('slope', slope)

  learning_rate = 0.1
  step_size = slope * learning_rate
  b3 = b3 - step_size
  print('step_size', step_size)
  print('new b3 =', b3)
  print('-'*100)
```

# Backpropagation (torch)

```python
import torch.optim as optim
import torch.nn.functional as F
import torch.nn as nn
import torch
```

```python
class Torch_Net(nn.Module):
  def __init__(self):
    super(Torch_Net, self).__init__()
    self.fc1 = nn.Linear(1, 2)
    self.fc2 = nn.Linear(2,1)

  def forward(self, x):
    x = F.softplus(self.fc1(x))
    x = self.fc2(x)
    return x
```

```python
input = torch.FloatTensor([[0],  
                          [0.5],  
                          [1]]).requires_grad_()
observed = torch.Tensor([[0],
                        [1],
                        [0]])

model = Torch_Net()

optimizer = optim.SGD(model.parameters(), lr=0.1)

for i in range(1,3000):
  pred = model(input)
  loss = F.mse_loss(pred, observed, reduction='sum')
  optimizer.zero_grad() 
  loss.backward() 
  optimizer.step()

  print('pred', pred)
  print("weights, biases", list(model.parameters()))
  print('loss', loss.item())
  print('-' * 50 + f'step_{i}' + '-' * 50)
  print()
```
